## Dataset Creation

```python
def fetch_trackids(user,playlist_id):
    trackids=[]
    playlist=s.user_playlist(user,playlist_id)
    for item in playlist['tracks']['items']:
        track=item['track']
        trackids.append(track['id'])
    return trackids
```

## Get Song Features

```python
def fetch_features(id):
    data=s.track(id)
    features=s.audio_features(id)
    
    name=data['name']
    album=data['album']['name']
    artist=data['album']['artists'][0]['name']
    release_date=data['album']['release_date']
    length=data['duration_ms']
    popularity=data['popularity']
    
    acousticness=features[0]['acousticness']
    danceability=features[0]['danceability']
    energy=features[0]['energy']
    instrumentalness=features[0]['instrumentalness']
    liveness=features[0]['liveness']
    loudness=features[0]['loudness']
    speechiness=features[0]['speechiness']
    tempo=features[0]['tempo']
    valence=features[0]['valence']
    time_signature=features[0]['time_signature']
    key=features[0]['key']
    mode=features[0]['mode']
    uri=features[0]['uri']
    
    track=[name,album,artist,release_date,length,popularity,
           acousticness,danceability,energy,instrumentalness,
          liveness,loudness,speechiness,tempo,valence,
          time_signature,key,mode,uri]
    return track
```

## Get All Playlist Song Details

```python
def fetch_songs(playlistids):
    tracks=[]
    for i in range(len(playlistids)):
        track=fetch_features(playlistids[i])
        tracks.append(track)
    return tracks
```

## Get Data and Put It Into DataFrame

```python
def fetch_data(tracks,year):
    df=pandas.DataFrame(tracks,columns=['name','album','artist','release_date','length','popularity',
           'acousticness','danceability','energy','instrumentalness',
          'liveness','loudness','speechiness','tempo','valence',
          'time_signature','key','mode','uri'])
    return df
```

```python
def fetch_years(df):
    years=[]
    for date in df['release_date'].values:
        if '-' in date:
            years.append(date.split('-')[0])
        else:
            years.append(date)
    df['release_year']=years
    return df
```
