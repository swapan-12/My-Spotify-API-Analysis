```python
#Loading EDA
import pandas
import numpy
import warnings
import time
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
workout=pandas.read_csv('workout_songs.csv')
punjabi=pandas.read_csv('punjabi_songs.csv')
```

```python
#Checking columns
workout.columns
```

```python
#Checking datatypes
workout.dtypes
```

```python
#Checking null values
workout.isnull().sum()
```

## Workout Playlist Each Artist Song Count

```python
plt.figure(figsize=(20,20))
sns.countplot(workout['artist'])
plt.xticks(rotation=90)
```

## Punjabi Playlist Each Artist Song Count

```python
plt.figure(figsize=(20,20))
sns.countplot(punjabi['artist'])
plt.xticks(rotation=90)
```

## Song Name by Popularity

```python
plt.figure(figsize=(16,16))
df=workout.groupby('name')['popularity'].mean().sort_values(ascending=False).head(10)
bar=sns.barplot(x=df.index,y=df)
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(16,16))
df=punjabi.groupby('name')['popularity'].mean().sort_values(ascending=False).head(10)
bar=sns.barplot(x=df.index,y=df)
plt.xticks(rotation=90)
```

```python
df_w=workout[['artist','name']].groupby('artist').count().sort_values('name',ascending=False).reset_index()
df_w.columns=['artist','count']
```

```python
df_w
```

```python
df_p=punjabi[['artist','name']].groupby('artist').count().sort_values('name',ascending=False).reset_index()
df_p.columns=['artist','count']
df_p
```

```python
df=pandas.concat([df_w,df_p],axis=1)
df
```
